```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Zadanie 1
def f(x):
    if np.fabs(x) < 1:
        return np.sin(1/x)
    else:
        return np.sin(x)
    
x = np.linspace(-10, 10, 200)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

```python
#Zadanie 2
def f(x):
    if x < 0:
        return 0
    else:
        return 1
    
x = np.linspace(-100, 100, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

```python
#Zadanie 3
def f(x):
    return np.power(x, 3) - np.power(x, 2)

x = np.linspace(-100, 100,1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

```python
x = np.linspace(-0.0001, 1.0001, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```
